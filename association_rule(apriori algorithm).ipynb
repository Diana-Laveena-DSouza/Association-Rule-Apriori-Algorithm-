```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
```

```python
#Load file
with open('Groceries.csv', 'r') as f:
    data=f.read()
    f.close()
dataset_new=list(data.split('\n'))
dataset=[]
for i in range(len(dataset_new)):
    dataset.append(list(dataset_new[i].lower().replace(' ,',',').split(',')))
dataset[0]
```

```python
#We can transform it into the right format via the TransactionEncoder as follows for 5 transactions
te=TransactionEncoder()
transaction_data=te.fit(dataset).transform(dataset)
df=pd.DataFrame(transaction_data, columns=te.columns_).drop('', axis=1)
df
```

```python
#Generating frequent items
frequent_items=apriori(df, min_support=0.006, use_colnames=True)
frequent_items['length']=frequent_items['itemsets'].apply(lambda x: len(x))
frequent_items
```

```python
#Association Rules
from apyori import apriori
association_rules=apriori(dataset, min_support=0.006, min_confidence=0.6)
for item in association_rules:
    print('Rule: '+str(item[2][0][0])+'->'+str(item[2][0][1]))
    print('support:'+str(item[1]))
    print('Confidence:'+str(item[2][0][2]))
    print('Lift: '+str(item[2][0][3]))
```

```python
#Changing to support=0.09, confidence=0.6
association_rules=apriori(dataset, min_support=0.009, min_confidence=0.6)
for item in association_rules:
    print('Rule: '+str(item[2][0][0])+'->'+str(item[2][0][1]))
    print('support:'+str(item[1]))
    print('Confidence:'+str(item[2][0][2]))
    print('Lift: '+str(item[2][0][3]))
```
